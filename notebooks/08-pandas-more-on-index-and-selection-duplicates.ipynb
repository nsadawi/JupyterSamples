```python
## get a list of column names
column_list = list(cars.columns)
column_list
```

```python
column_list = list(set(column_list))
```

```python
## reorder the columns of the DF according to the new ordered list
cars = cars[column_list]
cars
```

```python
## compute mean of rows and add it as a new column
## notice the axis = 1 .. 0 for columns
cars['row_avg'] = cars.mean(axis=1)
```

```python
## remove duplicate rows .. by default works on entire row
## possible to use a subset of the columns
cars.drop_duplicates(subset=['cyl'], keep='last')
```
