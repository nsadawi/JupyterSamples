## Attach dataframe as a new sheet to a new Excel file


```python
## Make sure you use a non-existing file name under that path
excel_writer = pd.ExcelWriter('/Users/csstnns/Desktop/Python Data Course - Oct2020/Day 2/UCL-updated_new.xlsx')
```

```python
## just to make things more inyteresting .. add new sheets
country_info = pd.read_csv('/Users/csstnns/Desktop/Python Data Course - Oct2020/Day 2/country_info.csv')
## create a pivot table
my_pivot_table = pd.pivot_table(country_info, index=['continent','country'], 
               values=['lifeExp','pop','gdpPercap'],
                                aggfunc=[np.std,np.median,'mean'])
```

```python
data.to_excel(excel_writer, sheet_name='Sheet3')
my_pivot_table.to_excel(excel_writer, sheet_name='CountryPivotTable')


excel_writer.close()
```
